### Preprocesssing Step

```python
data.isnull().sum()
```

```python
data.nunique()
```

```python
print("children : ",data['children'].unique())
print("sex : ",data['sex'].unique())
print("smoker : ",data['smoker'].unique())
print("region : ",data['region'].unique())
```

```python
data['sex'].replace({'female':0,'male':1},inplace=True)
data['smoker'].replace({'no':0,'yes':1},inplace=True)
data['region'].replace({'southwest': 0,'southeast':1, 'northwest':2, 'northeast':3},inplace=True)
```

```python
data
```

```python
X=data.drop('charges',axis =1)
y=data['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=56)
```

```python
X_train
```

```python

from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet,RidgeCV,LassoCV,ElasticNetCV

model_1=SVR()
model_1.fit(X_train,y_train)
model_1.score(X_test,y_test)*100
```

```python
def MLmodel(model,X_train,y_train,X_test,y_test):
    model_1=model()
    model_1.fit(X_train,y_train)
    print(model_1,'train score is :',model_1.score(X_train,y_train)*100)
    print(model_1,'test score is  :',model_1.score(X_test,y_test)*100)
    
```

```python
MLmodel(LinearRegression,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(Ridge,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(Lasso,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(ElasticNet,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(RidgeCV,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(LassoCV,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(SVR,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(DecisionTreeRegressor,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(ElasticNetCV,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(KNeighborsRegressor,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
MLmodel(RandomForestRegressor,X_train,y_train,X_test,y_test)
print('---------------------------------------------------------------------')
```
